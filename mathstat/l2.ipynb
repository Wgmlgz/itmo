```python
import matplotlib.pyplot as plt


def empirical_func(k):
    frequency_count = 0
    for j in sorted_data:
        if j < k:
            frequency_count += freqs[j]  # эмпирическая функция (вероятность того, что число из совокупности меньше, чем X)
        else:
            break
    return frequency_count / len(sorted_data)


data = [-4.22, -3.38, -3.04, -2.15, -5.53, -2.43, -4.00, -2.79,
        -2.75, -2.69, -3.67, -3.59, -2.45, -2.61, -3.98, -2.32]

sorted_data = sorted(data)

print(f"Вариационный ряд: {sorted_data}")

freqs = {x: data.count(x) for x in sorted_data}  # Словарь x: частота

f = [round(x / len(sorted_data), 4) for x in freqs.values()]  # Частоты делим на количество, получаем частотности
print("Частотности: ")
print(*f)

emp_values = [empirical_func(x) for x in sorted_data]
for i in range(len(sorted_data) - 1):
    plt.hlines(emp_values[i], sorted_data[i], sorted_data[i + 1])  # Считаем и рисуем эмпирическую функцию 
plt.title('Функция распределения')
plt.show()
```

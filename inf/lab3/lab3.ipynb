3. Программа должна считать количество смайликов определённого вида (вид смайлика описан в
таблице вариантов) в предложенном тексте. Все смайлики имеют такую структуру:
[глаза][нос][рот].
Вариантом является различные наборы глаз, носов и ртов.

Пример смайлика: 8<{P


```python
import re


def escapeRegex(s): return re.sub(r'([.*+?^${}()|[\]\\])', r'\\\1', s)

isu = 371364
eyes = [':', ';', 'X', '8', '=', '[']
noses = ['-', '<', '-{', '<{']
mouthes = ['(', ')', 'O', '|', '\\', '/', 'P']
smiley = eyes[isu % 6] + noses[isu % 4] + mouthes[isu % 7]
regex = escapeRegex(smiley)
print(f'{smiley=} {regex=}')


def test(value):
    print(len(re.findall(regex, value)))


test("asf][cX]a-]C]X}sfdasdkf{DFSc]:-(C}ZX}}D#}WE}R}ZXX")  # 1
test("asfdsfsdfsdf::;fsdfsd(((((((((((((((((")  # 0
test(":-(:::::::::::::((:-324(((_)_-------::())()()(:-()()")  # 2
test("!!!!^^^&#$&@#$4342432)_#(*$#*$@_#$@#*$+$#@$")  # 0
test("$@*:-((&@2423{|:-(}{|||{H(DS*F!@$3:-(42:-(4")  # 4
```

Хайку – жанр традиционной японской лирической поэзии века, известный с XIV века.
Оригинальное японское хайку состоит из 17 слогов, составляющих один столбец
иероглифов. Особыми разделительными словами – кирэдзи – текст хайку делится на
части из 5, 7 и снова 5 слогов. При переводе хайку на западные языки традиционно
вместо разделительного слова использую разрыв строки и, таким образом, хайку
записываются как трёхстишия.
Перед вами трёхстишия, которые претендуют на то, чтобы быть хайку. В качестве
разделителя строк используются символы «/». Если разделители делят текст на
строки, в которых 5/7/5 слогов, то выведите «Хайку!». Если число строк не равно 3,
то выведите строку «Не хайку. Должно быть 3 строки.». Иначе выведите строку вида
«Не хайку.»
Для простоты будем считать, что слогов ровно столько же, сколько гласных, не
задумываясь о тонкостях.

```python
def test(s):
    regex_strings = re.fullmatch(r'.+/.+/.+', s)
    x = r'([^аоуыэяёюие/]*[аоуыэяёюие][^аоуыэяёюие/]*)'
    regex = re.fullmatch(rf'{x}{{5}}\/{x}{{7}}\/{x}{{5}}', s, re.IGNORECASE)

    match(regex_strings is not None, regex is not None):
        case(0, _): print('Не хайку. Должно быть 3 строки.')
        case(1, 0): print('Не хайку.')
        case(1, 1): print('Хайку!')


test('Вечер за окном. / Еще один день прожит. / Жизнь скоротечна...')
test('Просто текст')
test('Как вишня расцвела! / Она с коня согнала / И князя-гордеца.')
test('Всё глазел на них, / Сакуры цветы, пока / Шею не свело')
test('Она с коня согнала / И князя-гордеца.')
```

# 3
Необходимо выбрать три любых буквы и расстояние между ними. С помощью
регулярного выражения нужно найти все слова (последовательность символов
ограниченная пробелами), в которых встречаются эти буквы в заданной
последовательности и расстояние (например, через один друг от друга).

```python
def test(s, letters=['К', 'Р', 'А'], dist=1):
    print(' '.join(re.findall(f'\w{{{dist}}}'.join(letters), s)))


test('КоРмА К Р А КоРмА')
test('КоРкА')
test('КоРчмА')
test('AbObA  КоРмА AbObuS', ['A', 'O', 'A'])
test('A___O666S', ['A', 'O', 'S'], 3)
test('КоР')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2 + x + np.sin(x)

def df(x):
    return 2*x + 1 + np.cos(x)

def ddf(x):
    return 2 - np.sin(x)

# Set parameters
a, b = -1, 0
eps = 10**-10
delta = eps
max_iter = 25

# Plot the function and its derivative
x_vals = np.linspace(a, b, 1000)
y_vals = f(x_vals)
dy_vals = df(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = $x^2 + x + sin(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

def plot(x, y, title):
    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_vals, label='f(x) = $x^2 + x + sin(x)$')
    plt.scatter(x, y, color='red', zorder=5, label=f'Minimum at x={x:.3f}, y={y:.3f}')  # Highlight the minimum
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()
    print(f"Minimum found at x = {x} with f(x) = {y}")
```

```python
def half_division(a, b, eps, max_iter, delta):
    iterations = []
    for i in range(max_iter):
        x1 = (a + b - delta) / 2
        x2 = (a + b + delta) / 2
        y1 = f(x1)
        y2 = f(x2)
        iterations.append([i+1, a, b, x1, y1, x2, y2])
        if y1 > y2:
            a = x1
        else:
            b = x2

        if b - a < 2 * eps:
            break
    x = (a + b) / 2
    y = f(x)
    return x, y, iterations


half_x, half_y, half_iterations = half_division(a, b, eps, max_iter, delta)

df_hd = pd.DataFrame(half_iterations, columns=[
                     'Iteration', 'a', 'b', 'x1', 'y1', 'x2', 'y2']).round(3)

plot(half_x, half_y, '  ')
df_hd
```

```python
def golden_ratio(a, b, eps, max_iter, delta):
    iterations = []
    phi = (1 + 5**0.5) / 2  # Golden ratio
    inv_phi = 1 -(1 / phi)
    
    x1 = a + inv_phi * (b - a)
    x2 = a + (1 - inv_phi) * (b - a)
    y1 = f(x1)
    y2 = f(x2)
    
    for i in range(max_iter):
        iterations.append([i+1, a, b, x1, y1, x2, y2])
        
        if y1 < y2:
            b = x2
            x2 = x1
            x1 = a + inv_phi * (b - a)
            y2 = y1
            y1 = f(x1)
        else:
            a = x1
            x1 = x2
            x2 = a + (1 - inv_phi) * (b - a)
            y1 = y2
            y2 = f(x2)
        
        if b - a < 2 * eps:
            break
    x = (a + b) / 2
    y = f(x)
    return x, y, iterations


golden_x, golden_y, golden_iterations = golden_ratio(a, b, eps, max_iter, delta)

df_hd = pd.DataFrame(golden_iterations, columns=[
                     'Iteration', 'a', 'b', 'x1', 'y1', 'x2', 'y2']).round(3)

plot(golden_x, golden_y, '  ')

df_hd
```

```python

def newton(a, b, eps, max_iter):
    iterations = []
    
    x = (a + b) / 2  # Initial guess
    for i in range(max_iter):
        y = f(x)
        iterations.append([i+1, x, y])
        
        x_new = x - df(x) / ddf(x)
        if abs(x_new - x) < eps:
            break
        x = x_new
    return x, y, iterations


newton_x, newton_y, newton_iterations = newton(a, b, eps, max_iter)

df_hd = pd.DataFrame(newton_iterations, columns=[
                     'Iteration', 'x', 'y']).round(3)
plot(newton_x, newton_y, ' ')

df_hd
```
